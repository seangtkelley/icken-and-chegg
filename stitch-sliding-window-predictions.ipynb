```python
import os

from PIL import Image
import PIL.ImageDraw as ImageDraw
import numpy as np
import time

from keras_yolo3.yolo import YOLO

annotation_path = './data/train.txt'
with open(annotation_path) as f:
    lines = f.readlines()

max_boxes = 20
      
detector = YOLO()

#for line in lines:
line = lines[0]
line = line.split()
image = Image.open(line[0])
iw, ih = image.size
h, w = (416, 416)
box = np.array([np.array(list(map(int,box.split(',')))) for box in line[1:]])

draw = ImageDraw.Draw(image)

current_window = 0
for y in range(0, ih, h):
    for x in range(0, iw, w):
        area = (x, y, x+w, y+h)
        window = image.crop(area)

        # predicted bounding boxes
        _, bounding_boxes = detector.detect_image_silent(window)

        for bb in bounding_boxes:
            draw.rectangle(
                [bb['left'] + x, bb['top'] + y, bb['right'] + x, bb['bottom'] + y],
                outline="red")

        # true bounding boxes
#             boxes_for_window = np.zeros((max_boxes,5))
#             if len(box)>0:
#                 boxes_added = 0
#                 for bb in box:
#                     if boxes_added < max_boxes and bounding_box_in_window(area, bb[:4]):

#                         # adjust bounding box coords to window
#                         adjusted_bb = [bb[0]-x, bb[1]-y, bb[2]-x, bb[3]-y, 0]

#                         # add box to list for this window
#                         boxes_for_window[boxes_added] = adjusted_bb
#                         boxes_added += 1

        current_window += 1
        
image.save('first_train_example_detections.tiff')
#break
```
